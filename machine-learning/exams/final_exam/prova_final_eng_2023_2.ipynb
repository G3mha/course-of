**Questão 3** (1,5 pts.): Faça uma pipeline que recebe um dataset `X_train` e retorna um `array` com os valores das duas principais componentes de X para cada amostra. Mostre um *scatter plot* destas componentes, onde a cor é `y_train`.

````python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Supondo que X_train e y_train já estejam carregados com os dados de treinamento
# Substitua X_train e y_train pelos nomes reais dos seus conjuntos de dados.

# Criar a pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Normalização dos dados
    ('pca', PCA(n_components=2))     # PCA com duas componentes principais
])

# Ajustar a pipeline aos dados de treinamento e transformar os dados
X_pca = pipeline.fit_transform(X_train)

# Criar o scatter plot
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_train)
plt.colorbar(scatter, label='y_train')
plt.title('Scatter Plot das Duas Principais Componentes com Cores Baseadas em y_train')
plt.xlabel('Primeira Componente Principal')
plt.ylabel('Segunda Componente Principal')
plt.show()
````

**Questão 4** (1,5 pt.): Faça uma pipeline que recebe um dataset `X` e usa `KMeans` para *clusterizar* o dataset em 7 clusters, ou seja, obter a alocação de cluster para cada amostra de entrada.

**ATENÇÃO**: ver questão 2!

<font color='red'>Utilizei ChatGPT para gerar o código da questão.</font>

````python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

import warnings
warnings.filterwarnings("ignore")

# Criar a pipeline
kmeans_pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Normalização dos dados
    ('kmeans', KMeans(n_clusters=7, random_state=42))  # KMeans com 7 clusters
])

# Ajustar a pipeline aos dados e obter as alocações de cluster para cada amostra
cluster_labels = kmeans_pipeline.fit_predict(X)

# Mostrar as alocações de cluster para cada amostra
print("Alocações de Cluster:")
print(cluster_labels)
````

````python
import warnings
warnings.filterwarnings("ignore")

# Ajustar a pipeline aos dados de treinamento e obter as alocações de cluster
X_train['cluster'] = kmeans_pipeline.fit_predict(X_train)

# Usar a pipeline treinada para obter as alocações de cluster para os dados de teste
X_test['cluster'] = kmeans_pipeline.predict(X_test)

# Verificar as primeiras linhas dos conjuntos de treinamento e teste para confirmar a adição da coluna 'cluster'
print("Conjunto de Treinamento com a nova coluna 'cluster':")
X_train.head()
````

**Questão 6** (1,5 pts.): Construa uma pipeline que processa os dados pós-Questão 5 de acordo com a seguinte especificação:

- Para a nova feature `cluster` use OneHotEncoding;
- Para as features originais, use uma sub-pipeline com 2 etapas:
    - `PolynomialFeatures` de grau 2;
    - `StandardScaler`;

Dica: use `ColumnTransformer`.

Processe os dados de treino e teste

<font color='red'>Utilizei ChatGPT para gerar o código da questão.</font>

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
import pandas as pd

# Definir as colunas para aplicar diferentes transformações
categorical_columns = ['cluster']
numeric_columns = X_train.columns.difference(['cluster']).tolist()

# Criar a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('cluster_onehot', OneHotEncoder(), categorical_columns),
        ('numeric_poly_scale', Pipeline([
            ('poly', PolynomialFeatures(degree=2, include_bias=False)),
            ('scale', StandardScaler())
        ]), numeric_columns)
    ])

# Construir a pipeline final
final_pipeline = Pipeline([
    ('preprocessor', preprocessor)
])

# Processar os dados de treinamento e teste
X_train_processed = final_pipeline.fit_transform(X_train)
X_test_processed = final_pipeline.transform(X_test)

# Converter os resultados de volta para DataFrames
X_train_processed_df = pd.DataFrame(X_train_processed, columns=final_pipeline.named_steps['preprocessor'].get_feature_names_out())
X_test_processed_df = pd.DataFrame(X_test_processed, columns=final_pipeline.named_steps['preprocessor'].get_feature_names_out())

# Verificar os resultados
print("Conjunto de Treinamento Processado:")
X_train_processed_df.head()
````

**Questão 7** (1,0 pt.): Avalie o desempenho de um modelo `LinearSVC` com coeficiente de regularização `C=1e2` nestes dados processados.

<font color='red'>Utilizei ChatGPT para gerar o código da questão.</font>

````python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

import warnings
warnings.filterwarnings("ignore")


# Criar e treinar o modelo LinearSVC
linear_svc_model = LinearSVC(C=1e2, random_state=42)
linear_svc_model.fit(X_train_processed, y_train)

# Fazer previsões nos dados de teste
y_pred = linear_svc_model.predict(X_test_processed)

# Avaliar a acurácia do modelo
accuracy = accuracy_score(y_test, y_pred)
print("Acurácia do modelo LinearSVC:", accuracy)
````

O modelo apresenta uma acurácia muito acima do trivial, 89% contra 22% do trivial, e portanto, parece ser um bom modelo para classificação.

A curva apresentada nos mostra que, à medida que o coeficiente de regularização C aumenta, a acurácia no conjunto de treinamento continua a crescer, enquanto a acurácia no conjunto de teste cai e posteriormente estabiliza. Este é um comportamento típico de overfitting.

Para decidir o que fazer agora que o modelo está sobreajustado, vamos entender melhor as características que compõe este modelo LinearSVC.

````python
transformed_features = final_pipeline.named_steps['preprocessor'].transform(X_train)
print("Número total de características após transformações:", transformed_features.shape[1])
````

````python
import numpy as np

coefficients = linear_svc_model.coef_
feature_importance = np.abs(coefficients).sum(axis=0)

# Visualizar a importância relativa das características
print("Importância relativa das características:", feature_importance)
````

Ao observar a importância de cada característica, podemos ver que muitas delas têm importância bem baixa, entre 2 e 3,99. Um passo a ser testado para melhorar o modelo é removê-las do conjunto de dados, pois elas podem estar atrapalhando o modelo a generalizar melhor.

Outra alternativa, é aumentar a força da regularização (C menor) para penalizar mais os coeficientes, evitando que ocorra este sobreajuste.

Sobre o desempenho pós-correção de sobreajuste, se a correção for eficaz, o desempenho no conjunto de teste pode melhorar, indicando uma melhor capacidade de generalização para dados não vistos.
No entanto, é importante estar atento para que também não ocorra o fênomeno oposto, o underfitting, onde o modelo se torna muito simples demais, não conseguindo capturar padrões nos dados de treinamento.
