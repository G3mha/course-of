 * $E$: batidas no lado esquerdo do goleiro
 * $R$: partidas em que ele se sentiu relaxado

**DEFINA AS PROBABILIDADES APRESENTADAS NO ENUNCIADO EM TERMOS DOS EVENTOS:**

 * $P(D)= 0.5$
 * $P(R|D)= 0.85$
 * $P(D|R)= 0.919$

**Deixe claro todo o racioc√≠nio usado para chegar na probabilidade questionada:**

```python
P_N = 1 - P_R
P_E = 1 - P_D


P_DeR = P_R*P_D_dado_R
P_DeN = P_D - P_DeR
P_D_dado_N = P_DeN / P_N
P_N_dado_D = P_N * P_D_dado_N / P_D

P_N_dado_D = P_N * P_D_dado_N / P_D
P_NeD = P_D*P_N_dado_D

P_NeE = P_N - P_NeD

P_E_dado_N = P_NeE / P_N
round(P_E_dado_N, 3)
```
